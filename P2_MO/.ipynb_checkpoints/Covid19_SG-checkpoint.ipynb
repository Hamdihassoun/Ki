```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
df = pd.read_csv('Covid-19_SG.csv')
df.set_index('Date', inplace=True)
df
```

```python
population = 5686000
#7-Tage-Inzidenz berechnen
df['7-Day Incidence'] = df['Daily Confirmed'].rolling(window=7).sum() * 100_000 / population
#Hospitalisierungsrate berechnen
df['Hospitalization Rate'] = df['Still Hospitalised'] / population * 100_000
#Kritikalitäts-Score berechnen
def calculate_criticality(incidence):
    if incidence > 400:
        return 4
    elif incidence > 200:
        return 3
    elif incidence > 100:
        return 2
    elif incidence > 50:
        return 1
    else:
        return 0
```

```python
df['Criticality Score'] = df['7-Day Incidence'].apply(calculate_criticality)
#  Visualisierungen erstellen
plt.figure(figsize=(15, 10))

# 4.1 7-Tage-Inzidenz
plt.subplot(3, 1, 1)
plt.plot(df.index, df['7-Day Incidence'], label='7-Tage-Inzidenz', color='blue')
plt.axhline(400, color='red', linestyle='--', label='Kritischer Schwellenwert (400)')
plt.title('7-Tage-Inzidenz in Singapur')
plt.xlabel('Datum')
plt.ylabel('Inzidenz (pro 100.000)')
plt.legend()
plt.grid()
```

```python
# 4.2 Hospitalisierungsrate
plt.subplot(3, 1, 2)
plt.plot(df.index, df['Hospitalization Rate'], label='Hospitalisierungsrate', color='green')
plt.title('Hospitalisierungsrate in Singapur')
plt.xlabel('Datum')
plt.ylabel('Rate (pro 100.000)')
plt.legend()
plt.grid()
```

```python
plt.subplot(3, 1, 3)
plt.plot(df.index, df['Criticality Score'], label='Kritikalitäts-Score', color='purple')
plt.axhline(2, color='yellow', linestyle='--', label='Kritischer Score (2)')
plt.title('Kritikalitäts-Score basierend auf 7-Tage-Inzidenz')
plt.xlabel('Datum')
plt.ylabel('Kritikalitäts-Score')
plt.legend()
plt.grid()
```

```python
# Beispiel: Anzahl täglicher Todesfälle visualisieren
plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Daily Deaths'], label='Tägliche Todesfälle', color='orange')
plt.title('Tägliche Todesfälle in Singapur')
plt.xlabel('Datum')
plt.ylabel('Anzahl Todesfälle')
plt.legend()
plt.grid()
plt.show()
```
